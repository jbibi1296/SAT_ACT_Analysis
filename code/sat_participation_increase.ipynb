#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code 
def std_dev_fxn (n):
    meen = np.sum(n)/len(n)
    summ = 0
    for i in n:
        summ += ((i-meen)**2)
    stddev = (summ/len(n))**.5
    return(stddev)
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where:

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns.T))**.5)# Makes sure you have enough rows
    ncols = nrows
    if len(list_of_columns.T) / ncols != ncols:
        if ncols == 1:
            ncols += 1
            nrows += 1
        else:
            ncols+=1
    fig, ax = plt.subplots(nrows, ncols) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    fig.subplots_adjust(wspace=1/nrows,hspace = 1/nrows)
    fig.set_figheight(nrows**2+10)
    fig.set_figwidth(ncols**2+10)
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        titlename = column.replace('_',' ')
        ax[i].hist(dataframe[column]) 
        ax[i].set_title(f'{titlename} Histogram', 
                        fontsize = 40 / nrows);
    if "Participation" in titlename:
        for axar in ax.flat:
            axar.set_xlabel(xlabel='Participation %',fontsize = 30/nrows) 
            axar.set_ylabel(ylabel='Count of states',fontsize = 30/nrows)
    elif "Participation" not in titlename:
        for axar in ax.flat:
            axar.set_xlabel(xlabel='Total/Composite Grade',fontsize = 30/nrows) 
            axar.set_ylabel(ylabel='Count of states',fontsize = 30/nrows)
# Hide x labels and tick labels for top plots and y ticks for right plots.
#     for axar in ax.flat:
#         axar.label_outer()
subplot_histograms(final,final.columns[1:])
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels

````python
# Code to make subplots of boxplots 
# 1 boxplot for each column in the dataframe
def subplot_boxplots(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns.T))**.5)# Makes sure you have enough rows and square roots it so we can assign the same value to the column field so it can print out close to a perfect square
    ncols = nrows
    if len(list_of_columns.T) / ncols != ncols: #For loop in case the number of columns in the dataframe does not have a perfect square
        if ncols == 1: #if only 1 row and 1 column, the add 1 more for each
            ncols += 1
            nrows += 1
        else:
            ncols+=1 
    fig, ax = plt.subplots(nrows, ncols) # fig is the plot and ax is the code referring to the plot
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    fig.subplots_adjust(wspace=1/nrows,hspace = 1/nrows) # add space in between the subplots to make it look cleaner
    fig.set_figheight(nrows**2+10)  # make the size of the figure bigger as the amount of rows and columns grow
    fig.set_figwidth(ncols**2+10)
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists, and the column name
        titlename = column.replace('_',' ') # take out the underscores from the column name and replaces with a space
        if "Participation" in titlename: # checks to see if the graph is regarding participation
            ax[i].set_xlabel(xlabel='Participation %',fontsize = 30/nrows) 
        else:
            ax[i].set_xlabel(xlabel='Score',fontsize = 30/nrows) 
        sns.boxplot(dataframe[column],ax=ax[i]) #calls the data into a boxplot and sets the ax to be in a subplot spot
        ax[i].set_title(f'{titlename} Distribution Boxplot', fontsize = 40 / nrows); # Sets the title of the graphs
subplot_boxplots(final,final.columns[1:])
````
